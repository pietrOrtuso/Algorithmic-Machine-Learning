````python
##########################################
############# DATA IMPORT ################
##########################################

house_data = pd.read_csv(base + 'train.csv',keep_default_na=False)

house_data[:10]
````

The dataset has 81 features, included the regressand one - SalePrice. Among the other 80, 79 (both numerical and categorical) describe different charachteristics of the houses, while the first one only represents the assigned ID. This has for sure no relation with the price, therefore we can discard it

````python
#extract the regressand column and remove it from the train data
prices = house_data.loc[:,"SalePrice"]
train_data = house_data.drop(["Id", "SalePrice"],1) #we also remove the useless ID
````

Not all the categorical features must be treated in the same way: the ordinal ones can be translated into numerical, integer value; the pure categorical ones, indeed, cannot be converted and has to be managed differently. For the first set, we provide a list of ordinal categorical features, extracting the corresponding values from the Data description.rtf file and calculating the values to replace in the data frame.

````python
headers = list(train_data)
i = 1
cat_dict = []

#list of ordinal categorical features
cat_ord_features = ["LotShape",
               "Utilities",
               "LandSlope",
               "ExterQual",
               "ExterCond",
               "BsmtQual",
               "BsmtCond",
               "BsmtExposure",
               "BsmtFinType1",
               "BsmtFinType2",
               "HeatingQC",
               "KitchenQual",
               "Functional",
               "FireplaceQu",
               "GarageFinish",
               "GarageQual",
               "GarageCond",
               "PoolQC"]

cat_ord_dict = {}
categories = []
with open(base + "Data description.rtf") as f:
    i = 0
    begin = False
    feature = ""
    for line in f.readlines()[11:]:
        
        #if the row starts with a slash, it either means a list of categories is starting or that it has come to and end
        #in case it is an end, we create the corresponding replace values array
        
        if line.split()[0] == "\\" and feature != "":
            begin = ~ begin
            if not begin:
                cat_ord_dict[feature].append(np.arange(len(cat_ord_dict[feature][0])-1,-1,-1))
                feature = ""   
                
        #if the feature is in our list (which has been created following the file order), we prepare to memorize its categories
        
        elif cat_ord_features[i] in line:
            feature = cat_ord_features[i]
            cat_ord_dict[feature] = [[]]
            if i < len(cat_ord_features) - 1:
                i += 1
                
        #appending the categories to the appropriate feature
        
        elif line.split()[0] != "\\" and begin and feature != "":
            cat_ord_dict[feature][0].append(line.split()[0].strip())
````

````python
for column in cat_ord_dict.keys():
    train_data[column] = train_data[column].replace(to_replace = cat_ord_dict[column][0], value = cat_ord_dict[column][1])
     
````

````python
exceptions = ['LotFrontage','MasVnrArea','GarageYrBlt']
for i in range(len(train_data.loc[0])):
    if headers[i] in exceptions:
        continue
    #we check through the first value if the data is numerical (included the replaced ordinal categorical) or categorical
    if type(train_data.loc[0][i]) is str:
        cat_dict.append(headers[i])

#now that we have distinguished numerical and categorical values, we split the original dataset
#this way, we can consider the two cases separately

train_data_cat = train_data[[x for x in cat_dict]]
train_data_num = train_data[[x for x in headers if x not in cat_dict]]

print("%d categorical columns" %len(list(train_data_cat)))
print("%d numerical columns" %len(list(train_data_num)))
````

Applying the above method, there were errors in type recogntion: LotFrontage, MasVnrArea and GarageYrBlt were infered by Pandas as a str instead of int. Therefore, they were managed manually to be included into the numerical dataset

````python
################################
######## DATA CLEANING  ########
################################
for column in headers:
    #mask = train_data
    if "NA" in str(train_data[column].unique()):
        print(column)    

train_data_num.LotFrontage = train_data_num.LotFrontage.replace(to_replace="NA",value = "0")
train_data_num.MasVnrArea = train_data_num.MasVnrArea.replace(to_replace="NA",value = "0")
train_data_num.GarageYrBlt = train_data_num.GarageYrBlt.replace(to_replace="NA",value = "2019")
````

````python
train_data_num.loc[:,'YearBuilt'] = 2019 - train_data_num.loc[:,'YearBuilt']
train_data_num.loc[:,'YearRemodAdd'] = 2019 - train_data_num.loc[:,'YearRemodAdd']

train_data_num['GarageYrBlt'] = train_data_num['GarageYrBlt'].astype(np.int32)
train_data_num.loc[:,'GarageYrBlt'] = 2019 - train_data_num.loc[:,'GarageYrBlt']

train_data_num.loc[:,'YrSold'] = 2019 - train_data_num.loc[:,'YrSold']
````

````python
#remove useless columns for the columns: the ones where for a given value of a column all the instances are almost there
for key in list(train_data_cat[:-1]):
    percentages = np.asarray(train_data_cat.groupby(key).size()/len(train_data_cat))
    for i in range(len(percentages)):
        if percentages[i]>0.93:
            train_data_cat = train_data_cat.drop(columns=key)
            break
````

````python
normalizer = RobustScaler()
df_normalized = pd.DataFrame(normalizer.fit_transform(train_data_num))
````

````python
#to perform regression with categorical values, it is necessary to dummy encode the categories
#if a feature has n categories (n as extracted by the training data), n binary features are created
#only one of these new features can be equal to 1 for each entry
house_data_cat_dummy =  pd.get_dummies(train_data_cat)

#rebuild the original dataset with all characteristics (categorical + numerical)
norm_data = pd.concat([house_data_cat_dummy,df_normalized], axis = 1)
````

Should we evaluate year? instead of having year, use something like age?

````python
x_train, x_test, y_train, y_test = train_test_split(norm_data, prices, test_size = 0.3)
model = LinearRegression()
model.fit(x_train,y_train)
y_predict = np.array(model.predict(x_test))
````

````python
from sklearn import metrics
print("Root Mean square error: " , np.sqrt(metrics.mean_squared_error(np.log(np.abs(y_test)),np.log(np.abs(y_predict)))))
results = pd.DataFrame()
results['TRUE'] = y_test
results['PREDICT'] = y_predict
````
